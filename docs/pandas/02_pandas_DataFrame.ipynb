```python
double_list = [
             [1000, '과자','2019-12-31','반품'],
             [2000, '음료', '2020-03-02','정상'],
             [3000, '아이스크림', '2020-02-03','정상'],
             [1000,'과자','2019-12-31','반품']
            ]
pd.DataFrame(data=double_list)
```

## 전처리

### 추출

#### 컬럼 단위

```python
df_saledays.columns
```

```python
# df_saledays['가격'] # serise 추출
serise_price = df_saledays['가격'] # serise 추출
serise_price.shape, type(serise_price)
```

```python
# df_saledays[['종류','판매일자']]  # dataframe 추출
df_types = df_saledays[['종류','판매일자']]  # dataframe 추출
type(df_types), df_types.value_counts()
```

#### row(index) 단위 :
- iloc : 위치값으로 접근
- loc : 명칭 접근

```python
df_saledays.index = ['a', 'b', 'c', 'd']
```

```python
df_saledays
```

```python
# df_saledays.iloc[2]
df_saledays.loc['c']
```

#### index와 columns 같이 추출

```python
df_saledays.loc['c', '판매일자']  # indexes, columns
```

```python
df_saledays.loc[:, '판매일자']  # indexes, columns
```

```python
df_saledays.columns
```

```python
df_saledays.loc[:, '종류':]  # indexes, columns
```

### 정렬

```python
df_saledays.sort_values(by='가격', ascending=True)  # 단일 변수일 때
```

```python
df_saledays.sort_values(by=['가격', '반품여부'], ascending=[False, True])  # 다변수일 때
```
